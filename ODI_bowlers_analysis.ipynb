###### 8. How many players played for ICC?

```python
def icc_check(x):
    if 'ICC' in x:
        return 'Yes'
    else:
        return 'No'
```

```python
df['played_for_ICC'] = df['Player'].apply(icc_check)
```

```python
display(df.head())
```

```python
print(df['played_for_ICC'].value_counts())
```

###### 13 players played for ICC.

###### 10. Which player(s) had played for the longest period of time?

```python
df['Start_year'] = df['Playing_span'].str[:4]
df['End_year'] = df['Playing_span'].str[5:]
```

```python
display(df.head())
```

```python
df=df.drop('Playing_span', axis=1)
```

```python
df.head()
```

```python
df.info()
```

```python
df['Start_year'] = df['Start_year'].astype('int') 
df['End_year'] = df['End_year'].astype('int')
```

```python
df.head()
```

```python
df['Years_played'] = df['End_year'] - df['Start_year']+1
```

```python
df.head()
```

```python
df = df.drop(['Start_year','End_year'], axis=1)
```

```python
df.head()
```

```python
df.sort_values(by='Years_played', ascending = False).head()
```

###### From the sorted list above SR Tendulkar had played for the longest period of time, 24 years.

###### 11. Which player(s) had played for the shortest period of time?
