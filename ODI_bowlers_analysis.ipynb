```python
df = df.rename(columns={'Span':'Playing_span',
                        'Mat':'Matches_played',
                        'Inns':'Innings_bowled',
                        'Balls':'Balls_bowled',
                        'Runs':'Runs_conceded',
                        'Wkts':'Wickets_taken',
                        'Ave':'Bowling_average',
                        'Econ':'Economy_rate',
                        'SR':'Bowling_strike_rate',
                        '4':'four_wkts_in_an_inns',
                        '5':'five_wkts_in_an_inns'})
```

```python
display(df.head())
```

```python
# 8. How many players played for ICC?
```

```python
def icc_check(x):
    if "ICC" in x:
        return "Yes"
    else:
        return "No"
```

```python
df['played_for_ICC'] = df['Player'].apply(icc_check)
```

```python
display(df.head())
```

```python
print(df['played_for_ICC'].value_counts())
```

```python
# 9. How many different countries are present in this dataset? 
df[["Player_Name", "Country"]] = df['Player'].str.split("(", expand=True)
```

```python
display(df.head())
```

```python
df.drop('Player', axis=1, inplace=True)
```

```python
display(df.head())
```

```python
df['Country'] = df['Country'].str.replace(")", "")
```

```python
display(df.head())
```

```python
print(df['Country'].value_counts())
```

```python
def Asia_check(x):
    if "Asia" in x:
        return "Yes"
    else:
        return "No"
```

```python
def Afr_check(x):
    if "Afr" in x:
        return "Yes"
    else:
        return "No"
```

```python
df['played_for_Asia'] = df['Country'].apply(Asia_check)
df['played_for_Afr'] = df['Country'].apply(Afr_check)
```

```python
display(df.head())
```

```python
display(df.columns)
```

```python
new_col_sequence = ['Player_Name', 'Country', 'Playing_span', 'Matches_played', 'Innings_bowled', 'Balls_bowled',
       'Runs_conceded', 'Wickets_taken', 'Bowling_average', 'Economy_rate',
       'Bowling_strike_rate', 'four_wkts_in_an_inns', 'five_wkts_in_an_inns',
       'played_for_ICC', 'played_for_Asia',
       'played_for_Afr']
```

```python
display(df.head())
```

```python
df = df[new_col_sequence]
```

```python
display(df.head())
```

```python
print(df['Country'].value_counts())
```

```python
df['Country'] = df['Country'].str.replace("Asia", "")
df['Country'] = df['Country'].str.replace("ICC", "")
df['Country'] = df['Country'].str.replace("Afr", "")
df['Country'] = df['Country'].str.replace("/", "")
```

```python
print(df['Country'].value_counts())
```

```python
# 10. Which player(s) had played for the longest period of time?
df['Start_year'] = df['Playing_span'].str[:4]
df['End_year'] = df['Playing_span'].str[5:]
```

```python
display(df.head())
```

```python
df=df.drop('Playing_span', axis=1)
```

```python
df
```

```python
display(df.info)
```

```python
df.info()
```

```python
df['Start_year'] = df['Start_year'].astype('int') 
df['End_year'] = df['End_year'].astype('int')
```

```python
df
```

```python
df['Years_played'] = df['End_year'] - df['Start_year']+1
```

```python
df
```

```python
df = df.drop(['Start_year','End_year'], axis=1)
```
